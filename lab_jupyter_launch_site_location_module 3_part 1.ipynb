```python
# TASK 3: Calculate the distances between a launch site to its proximities
from math import radians, cos, sin, asin, sqrt

def calculate_distance(lat1, lon1, lat2, lon2):
    # Convert degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    
    # Haversine formula
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 6371  # Radius of Earth in kilometers
    return c * r
proximities = {
    'Nearest City': [28.3922, -80.6077],      # Example: Cocoa Beach, FL
    'Nearest Coastline': [28.5641, -80.5678], # Example: Cape Canaveral coast
    'Nearest Highway': [28.5721, -80.6480],   # Example: Highway intersection
    'Nearest Railway': [28.573255, -80.646895]
}
for index, site in launch_sites.iterrows():
    print(f"Distances from {site['Launch Site']}:")
    for name, (lat, lon) in proximities.items():
        dist = calculate_distance(site['Lat'], site['Long'], lat, lon)
        print(f"  To {name}: {dist:.2f} km")
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# find coordinate of the closet coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163
# Sample coordinates of coastlines (add more if needed)
coastlines = [
    [28.5641, -80.5678],  # Near CCAFS
    [28.5500, -80.6000],  
    [28.5700, -80.5800]
]
# Example launch site coordinate (CCAFS LC-40)
launch_lat = 28.562302
launch_lon = -80.577356
# Initialize minimum distance and closest point
min_dist = float('inf')
closest_coast = None

# Loop over each coastline point
for coast_lat, coast_lon in coastlines:
    dist = calculate_distance(launch_lat, launch_lon, coast_lat, coast_lon)
    if dist < min_dist:
        min_dist = dist
        closest_coast = (coast_lat, coast_lon)

# Output the result
print(f"Closest coastline coordinate: {closest_coast}")
print(f"Distance: {min_dist:.2f} km")

```

```python
launch_lat, launch_lon = 28.562302, -80.577356  # Example launch site

# Sample closest features (you can replace these with actual values)
city_coord = [28.6100, -80.6040]
railway_coord = [28.5700, -80.5700]
highway_coord = [28.5900, -80.5900]
```

```python
import folium

# Create map
site_map = folium.Map(location=[launch_lat, launch_lon], zoom_start=12)

# Add launch site marker
folium.Marker(
    location=[launch_lat, launch_lon],
    popup="Launch Site",
    icon=folium.Icon(color='blue')
).add_to(site_map)

# Helper function to add a marker and line
def add_proximity_marker_and_line(feature_name, feature_coord, color):
    dist = calculate_distance(launch_lat, launch_lon, feature_coord[0], feature_coord[1])
    
    # Marker
    folium.Marker(
        location=feature_coord,
        popup=f"{feature_name}\nDistance: {dist:.2f} km",
        icon=folium.Icon(color=color, icon='info-sign')
    ).add_to(site_map)
    
    # Line
    folium.PolyLine(
        locations=[[launch_lat, launch_lon], feature_coord],
        color=color,
        weight=2,
        tooltip=f"{feature_name} distance: {dist:.2f} km"
    ).add_to(site_map)

# Add proximity features
add_proximity_marker_and_line("City", city_coord, "green")
add_proximity_marker_and_line("Railway", railway_coord, "orange")
add_proximity_marker_and_line("Highway", highway_coord, "purple")

# Display map
site_map
```
